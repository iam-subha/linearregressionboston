```python
#Now we fetch some data and store it as a test data so that we can use it after the training of the model to check the performance of the model
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
print(x.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#Standardize the dataset
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#Now we use stratifiedshufflesplit to equaly distribute the data in the train and test set
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
sp=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in sp.split(dataset,dataset["CHAS"]):
    st_train_data=dataset.loc[train_index]
    st_test_data=dataset.loc[test_index]
```

```python
from sklearn.linear_model import LinearRegression
lrg=LinearRegression()
lrg.fit(x_train,y_train)
print(lrg.coef_)
print(lrg.intercept_)
```

```python
##prediction with the test data
pred_lrg=lrg.predict(x_test)
```

```python
print(pred_lrg)
```

```python
plt.scatter(y_test,pred_lrg)
```

```python
##check the performance of the model
from sklearn.metrics import mean_squared_error
print (mean_squared_error(y_test,pred_lrg))
print(np.sqrt(mean_squared_error(y_test,pred_lrg)))
## check R square 
from sklearn.metrics import r2_score
print(r2_score(y_test,pred_lrg))
```
